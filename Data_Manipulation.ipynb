### Exercises (MANDATORY CONCEPTS)

#### 🔹 Merging

##### 1. Perform inner join using:
* `pd.merge()`

````python
# import pandas as pd

df3 = pd.DataFrame({
    "EmpID": [1, 2, 3],
    "Name": ["Ali", "Sara", "Ahmed"]
})

df4 = pd.DataFrame({
    "EmpID": [1, 2, 4],
    "Salary": [50000, 60000, 70000]
})
# pd.merge(df3, df4, on="EmpID", how="inner") # how="inner" is default
pd.merge(df3, df4, on="EmpID")
````

##### 2. Perform left join

````python
df5 = pd.merge(df3, df4, on="EmpID", how="outer")
df5
````

##### 3. Identify missing matches

````python
missing_matches = df5[df5['Name'].isna() | df5['Salary'].isna()]
print(f"The missing matches in the merged DataFrame are: \n{missing_matches}")
````

##### 5. Convert wide → long format:

````python
df.melt(id_vars=['Name'], var_name='Category', value_name='Value')
````

###  🔴 LEVEL 5: Advanced Data Analysis

### Exercises

#### 1. Create a column:
* `Salary Category (Low, Medium, High)`

````python
df.loc[:,'Salary_Category'] = pd.qcut(df['Salary'], q=3, labels=['Low', 'Medium', 'High'])
````

````python
df
````

#### 2. Apply custom function using apply()

````python
df['Age_category'] = df['Age'].apply(lambda x: 'Young' if x < 25 else ('Middle_Aged' if x < 35 else 'Old'))
df
````

#### 3. Rank employees by salary

````python
df['Salary_Rank'] = df['Salary'].rank(ascending=False)
df
````

#### 4. Detect outliers using IQR

````python
# Calculate Q1, Q3, and IQR
Q1 = df['Salary'].quantile(0.25)
Q3 = df['Salary'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter outliers
filtered_df = df[(df['Salary'] >= lower_bound) & (df['Salary'] <= upper_bound)]
````

````python
filtered_df
````

#### 5. Normalize salary column

```
# Min-Max Normalization (Scale to 0-1)
(value - min) / (max - min)
```

````python
# df['column_name'] = (df['column_name'] - df['column_name'].min()) / (df['column_name'].max() - df['column_name'].min())
df['Normalized_Salary'] = (df['Salary'] - df['Salary'].min()) / (df['Salary'].max() - df['Salary'].min())
df
````

### 🔴 LEVEL 5: Linear Algebra (Professional 🔥)

### Exercises

#### 1. Matrix multiplication

````python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
# result = A @ B
result = np.matmul(A,B)
result
````

#### 2. Determinant

````python
# Calculate determinant

det = np.linalg.det(A)
print(f"Determinant: {det}")
````

#### 3. Inverse

````python
# 1. Matrix Inversion
inv_A = np.linalg.inv(A)
print(f"Inverse of A:\n{inv_A}\n")

# 2. Bit-wise Inversion
bit_inv_A = np.bitwise_not(A)
print(f"Bit-wise Inverse of A:\n{bit_inv_A}\n")

# Reversing Elements (Order)
reversed_A = np.flip(A)
print(f"Reversed A:\n{reversed_A}")
````

#### 4. Eigenvalues & eigenvectors

````python
# Eigenvalues
eigenvalues, eigenvectors = np.linalg.eig(A)
print("Eigenvalues:")
print(eigenvalues)
print("\nEigenvectors:")
print(eigenvectors)
````
